### 2. 合并数据集
- 使用相同的 id 合并 messages 和 categories 数据集
- 将合并后的数据集传入 `df`，在后续的步骤中要清洗该数据集。

```python
# merge datasets
#TODO 5
#可以使用pandas的内置函数merge来合并数据
df = pd.merge(messages,categories,on='id')
df.head(-1)
```

```python
df.shape
```

```python
df.head(1)
```

### 3. 分割 `categories` 
- 将 `categories` 列的值根据字符 `;` 进行分割，每个值是一个新列。你会发现 [这个方法](https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.Series.str.split.html) 非常有用。确保设置 `expand=True`。
- 使用 categories 数据框的第一行来创建类别数据的列名。
- 使用新的列名重命名 `categories` 的列。

```python
# create a dataframe of the 36 individual category columns
# TODO 6
categories = df['categories'].str.split(";", expand=True)
categories.head()
```

```python
row_my = categories.iloc[0]
# print(row_my)
cate_column = list(map(lambda x: x.split('-')[0], row_my))
cate_column
```

```python
# select the first row of the categories dataframe
# TODO 7
row = categories.iloc[0]

# use this row to extract a list of new column names for categories.
# one way is to apply a lambda function that takes everything 
# up to the second to last character of each string with slicing
category_colnames = list(map(lambda x: x.split('-')[0], row))
print(category_colnames)
```

```python
# rename the columns of `categories`
# TODO 8
categories.columns = category_colnames
categories.head()
```

### 4. 转换类别值至数值 0 或 1
- 遍历 df 的类别列，只保留每个字符串的最后一个字符（1 或 0）。例如，`related-0` 变成 `0`，`related-1` 变成 `1`。将字符串转换成数值。
- 通过在 Series 后加上`.str`，你可以实现 [Pandas Series 上的常见字符串操作](https://pandas.pydata.org/pandas-docs/stable/text.html#indexing-with-str)，如索引。你可能需要先将 Series 转换成类型字符串，你可以通过 `astype(str)` 实现。

```python
categories.shape
```

```python
for column in categories:
    categories[column] = categories[column].str.split('-').str[1]
#     print(categories[column])
categories.head()

```

```python
categories.head()
```

```python
for column in categories:
    categories[column] = categories[column].astype("int")
```

```python
categories.head()
```

```python
# TODO 9
for column in categories:
    # set each value to be the last character of the string
    #Elements in the split lists can be accessed using get or [] notation
    categories[column] = categories[column].str.split('-').str[1] 
    
    # convert column from string to numeric
    categories[column] = categories[column].astype("int")
categories.head()
```

```python
categories.head(-1)
```

```python
df.head(-1)
```

### 5. 替换 `df` `categories` 类别列
- 从 df 数据框中删除类别列，因为不再需要类别列了。
- 将 df 和 categories 数据框连接起来。

```python
# drop the original categories column from `df`
# TODO 10
df = df.drop(columns='categories', axis=1)

df.head()
```

```python
# test_df = df
test_df = pd.concat([df,categories], axis=1)
test_df.head()
```

```python
# concatenate the original dataframe with the new `categories` dataframe
# TODO 11
# 在这里进行合并，应该使用join，
# 在索引或键列上将列与其他DataFrame连接起来。通过传递一个列表，一次有效地通过索引连接多个DataFrame对象。
df = pd.concat([df,categories], axis=1)
df.head()
```

### 6. 删除重复行
- 检查这个数据集中有多少重复行。
- 删除重复行。
- 确认重复行被删除了。

```python
# check number of duplicates
# TODO 12
# duplicated_row = 
flag = df.duplicated()
duplicated_row = len([i for i in df.duplicated() if i])
duplicated_row
```

```python
# drop duplicates
# TODO 13
df = df.drop_duplicates(subset=None, keep='first', inplace=False)
df.head()
```

```python
# check number of duplicates
# TODO 14
flag = df.duplicated()
duplicated_row = len([i for i in flag if i])
duplicated_row
```

### 7. 保存整理后的数据集为 sqlite databse 
你可以使用 pandas 的 [`to_sql` 方法](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.to_sql.html) 和SQLAlchemy 库实现这一点。记得在 notebook 的第一个单元格导入 SQLAlchemy 的 `create_engine` 以在下方使用该库。

```python
# TODO 15
engine = create_engine('sqlite:///../data/udecity_project.db')

df.to_sql('DisasterResponsePipeline_table', engine, index=False)
# engine = create_engine('sqlite:///InsertDatabaseName.db')
# df.to_sql('InsertTableName', engine, index=False)
```
